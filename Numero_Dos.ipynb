```python
teams = teams.dropna()
```

```python
teams
```

```python
teams_train_data = teams[teams["year"] < 2012].copy()
teams_test_data = teams[teams["year"] >= 2012].copy()
```

```python
teams_train_data.shape
```

```python
teams_test_data.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
predictors = ["athletes", "prev_medals"]
```

```python
target = "medals"
```

```python
reg.fit(teams_train_data[predictors], teams_train_data["medals"])
```

```python
predictions = reg.predict(teams_test_data[predictors])
```

```python
predictions
```

```python
teams_test_data["predictions"] = predictions
teams_test_data
```

```python
teams_test_data.loc[teams_test_data["predictions"] < 0, "predictions"] = 0
```

```python
teams_test_data["predictions"] = teams_test_data["predictions"].round()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
error = mean_absolute_error(teams_test_data["medals"], teams_test_data["predictions"] )
```
